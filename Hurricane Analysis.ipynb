3. In addition to organizing the hurricanes in a dictionary with names as the key, you want to be able to organize the hurricanes by year.

   Write a function that converts the current dictionary of hurricanes to a new dictionary, where the keys are years and the values are lists containing a dictionary for each hurricane that occurred in that year.
   
   For example, the key `1932` would yield the value: `[{'Name': 'Bahamas', 'Month': 'September', 'Year': 1932, 'Max Sustained Wind': 160, 'Areas Affected': ['The Bahamas', 'Northeastern United States'], 'Damage': 'Damage not recorded', 'Deaths': 16}, {'Name': 'Cuba II', 'Month': 'November', 'Year': 1932, 'Max Sustained Wind': 175, 'Areas Affected': ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], 'Damage': 40000000.0, 'Deaths': 3103}]`.
   
   Test your function on your hurricane dictionary.

````python
# 3
# Organizing by Year
def organize_by_year(hurricanes):
    hurricanes_by_year = {}
    for hurricane in hurricanes.values():
        year = hurricane['Year']
        if year not in hurricanes_by_year:
            hurricanes_by_year[year] = []
        hurricanes_by_year[year].append(hurricane)
    return hurricanes_by_year
# create a new dictionary of hurricanes with year and key

hurricanes_by_year = organize_by_year(hurricanes)
hurricanes_by_year
````

4. You believe that knowing how often each of the areas of the Atlantic are affected by these strong hurricanes is important for making preparations for future hurricanes.

   Write a function that counts how often each area is listed as an affected area of a hurricane. Store and return the results in a dictionary where the keys are the affected areas and the values are counts of how many times the areas were affected.
   
   Test your function on your hurricane dictionary.

````python
# 4
# Counting Damaged Areas

# create dictionary of areas to store the number of hurricanes involved in
def count_affected_areas(hurricanes):
    affected_areas = {}
    for hurricane in hurricanes.values():
        for area in hurricane['Areas Affected']:
            if area not in affected_areas:
                affected_areas[area] = 0
            affected_areas[area] += 1
    return affected_areas

# create a new dictionary of areas affected by hurricanes
affected_areas = count_affected_areas(hurricanes)
affected_areas
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Order the affected areas by the number of hurricanes in plot
def order_affected_areas(affected_areas):
    return dict(sorted(affected_areas.items(), key=lambda item: item[1], reverse=True))
affected_areas = order_affected_areas(affected_areas)

sns.barplot(x=list(affected_areas.keys()), y=list(affected_areas.values()))
plt.xticks(rotation=90)
plt.title('Number of Hurricanes by Affected Area')
plt.xlabel('Affected Area')
plt.ylabel('Number of Hurricanes')
plt.show()
````

5. Write a function that finds the area affected by the most hurricanes, and how often it was hit.

   Test your function on your affected area dictionary.

````python
# 5
# Calculating Maximum Hurricane Count

# find most frequently affected area and the number of hurricanes involved in and how often it was affected
def max_hurricane_count(affected_areas):
    max_area = max(affected_areas, key=affected_areas.get)
    max_count = affected_areas[max_area]
    return max_area, max_count
max_area, max_count = max_hurricane_count(affected_areas)
print(f"The most frequently affected area is {max_area} with {max_count} hurricanes involved")
````

6. Write a function that finds the hurricane that caused the greatest number of deaths, and how many deaths it caused.

   Test your function on your hurricane dictionary.

````python
# 6
# Calculating the Deadliest Hurricane

# find highest mortality hurricane and the number of deaths
def deadliest_hurricane(hurricanes):
    max_deaths = 0
    deadliest = None
    for hurricane in hurricanes.values():
        if hurricane['Deaths'] > max_deaths:
            max_deaths = hurricane['Deaths']
            deadliest = hurricane
    return deadliest, max_deaths
deadliest, max_deaths = deadliest_hurricane(hurricanes)
print(f"The deadliest hurricane is {deadliest['Name']} with {max_deaths} deaths.")
````

7. Just as hurricanes are rated by their windspeed, you want to try rating hurricanes based on other metrics.

   Write a function that rates hurricanes on a mortality scale according to the following ratings, where the key is the rating and the value is the upper bound of deaths for that rating.
   
   ```py
   mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
   ```
   
   For example, a hurricane with a `1` mortality rating would have resulted in greater than `0` but less than or equal to `100` deaths. A hurricane with a `5` mortality would have resulted in greater than `10000` deaths.
   
   Store the hurricanes in a new dictionary where the keys are the mortaility ratings and the values are lists containing a dictionary for each hurricane that falls into that mortality rating.
   
   Test your function on your hurricane dictionary.

````python
# 7
# Rating Hurricanes by Mortality
mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
# categorize hurricanes in new dictionary with mortality severity as key
def categorize_hurricanes_by_mortality(hurricanes):
    categorized_hurricanes = {}
    for hurricane in hurricanes.values():
        mortality = hurricane['Deaths']
        if mortality < 100:
            category = 0
        elif mortality < 500:
            category = 1
        elif mortality < 1000:
            category = 2
        elif mortality < 10000:
            category = 3
        else:
            category = 4
        
        if category not in categorized_hurricanes:
            categorized_hurricanes[category] = []
        categorized_hurricanes[category].append(hurricane)
    
    return categorized_hurricanes
# create a new dictionary of hurricanes categorized by mortality severity
categorized_hurricanes = categorize_hurricanes_by_mortality(hurricanes)
categorized_hurricanes
````

8. Write a function that finds the hurricane that caused the greatest damage, and how costly it was.

   Test your function on your hurricane dictionary.

````python
# 8
# Calculating Hurricane Maximum Damage

# find highest damage inducing hurricane and its total cost
def max_damage_hurricane(hurricanes):
    max_damage = 0
    max_hurricane = None
    for hurricane in hurricanes.values():
        damage_str = hurricane['Damages (USD($))']
        if damage_str != 'Damages not recorded':
            if 'M' in damage_str:
                damage = float(damage_str.replace('M', '').replace('$', '').strip()) * 1000000
            elif 'B' in damage_str:
                damage = float(damage_str.replace('B', '').replace('$', '').strip()) * 1000000000
            else:
                continue
            
            if damage > max_damage:
                max_damage = damage
                max_hurricane = hurricane
    
    return max_hurricane, max_damage
max_hurricane, max_damage = max_damage_hurricane(hurricanes)
print(f"The hurricane with the highest damage is {max_hurricane['Name']} with a total cost of ${max_damage:,.2f}.")
````

````python
#  Plot the hurricane damages
def plot_hurricane_damages(hurricanes):
    damages = []
    names = []
    for hurricane in hurricanes.values():
        damage_str = hurricane['Damages (USD($))']
        if damage_str != 'Damages not recorded':
            if 'M' in damage_str:
                damage = float(damage_str.replace('M', '').replace('$', '').strip()) * 1000000
            elif 'B' in damage_str:
                damage = float(damage_str.replace('B', '').replace('$', '').strip()) * 1000000000
            else:
                continue
            
            damages.append(damage)
            names.append(hurricane['Name'])
    
    sns.barplot(x=names, y=damages)
    plt.xticks(rotation=90)
    plt.title('Hurricane Damages')
    plt.xlabel('Hurricane Name')
    plt.ylabel('Damage (USD($))')
    plt.show()

plot_hurricane_damages(hurricanes)
````
